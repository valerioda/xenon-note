# Acceptance calculation

```python
ev_sim = np.load('/dali/lgrandi/shenyangshi/wfsim_0530.npy', allow_pickle=True)
ev_sim = ev_sim[FiducialVolume.cut_by(ev_sim)]
NearWires = st.get_single_plugin('030000', 'cut_near_wires')
mask_near_wires = NearWires.cut_by(ev_sim)
```

```python
quantile_par = (( 0.63734546, 2.99678884, 1.03246656 ), ( 4.43451121, 1.84825809, 2.15736571, 1.2112417 ) )
parabola_par = ((-0.00798012,  0.23778128, -0.1894035), (0.11170658, -1.12877236,  4.10119736) )
title = 'WFsim - FarWires 1-99%'
evc =  ev_sim[~mask_near_wires]
cut = wcu.cut_s2_wdith_sim(evc,quantile_par,parabola_par, arange = (2.2,4.5),wrange =(0,5),
                     wires='far',drange=(-5000,25000), wire_model=0, title=title)
data = wcu.get_acceptance(evc, cut, title=title, bins = 40, alim=(0.9,1))
data.to_hdf(f's2widthcut_acceptance.h5', key='df', mode='w')
```

```python
data = pd.read_hdf(f's2widthcut_acceptance.h5')
param_list = ('s1_area','s2_area','z')
col_list = ('g','r','m')
for i, param_name in enumerate(param_list):
    plt.figure(figsize=(12,6))
    plt.errorbar(data[param_name], data[param_name+'_acc'],
                 yerr=(data[param_name+'_acc_err_l'], data[param_name+'_acc_err_h']),
                 color=col_list[i],fmt="s", ms=2, capsize=3,label='acceptance vs '+param_name)
    plt.xlabel(param_list[i], ha='right', x=1,fontsize=12)
    plt.ylabel("S2WidthCut Acceptance", ha='right', y=1,fontsize=12)
    plt.ylim(0.9,1)
    plt.title(f'{title} Total Acceptance = {acc_mean:.2f}%',fontsize=16)
```
