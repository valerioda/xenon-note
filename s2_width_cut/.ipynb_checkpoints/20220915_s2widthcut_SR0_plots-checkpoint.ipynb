```python
def save_boundaries(name=None):    
    area_space=(2,7)
    bins=300
    aspace = np.linspace(2, 7, bins)
    hh = int(2/5*bins)-1
    ac = (aspace[1:]+aspace[:-1])/2
    bound_low = np.concatenate((wcu.f_lower_boundary(ac[:hh],*fit_par[0]),
                                wcu.parabola(ac[hh:],*par_par[0]) ))
    bound_upp = np.concatenate((wcu.f_upper_boundary(ac[:hh],*fit_par[1]),
                                wcu.parabola(ac[hh:],*par_par[1]) ))
    low_bound = interp1d(ac, bound_low, bounds_error=False,fill_value='extrapolate',
                         kind='cubic')
    upp_bound = interp1d(ac, bound_upp, bounds_error=False,fill_value='extrapolate',
                         kind='cubic')
    data = pd.DataFrame(columns=['s2arealog','low_boundary','up_boundary'])
    data['s2arealog'] = ac
    data['low_boundary'] = low_bound(ac)
    data['up_boundary'] = upp_bound(ac)
    if name is not None: data.to_pickle(name)
    return data
```

```python
data = save_boundaries(name=f'{ddir}/SR0_v8_S2WidthCut_LowER.pkl')
```

```python
def save_boundary_wire_model(name=None):
    data = pd.DataFrame(columns=['x2w',
                                 'uppper_lim_with_wire_1',
                                 'uppper_lim_with_wire_2',
                                 'uppper_lim_with_wire_3',
                                 'uppper_lim_with_wire_4'])
    x2w = np.linspace(-10, 10, 1001)
    data['x2w'] = x2w
    x1 = np.ones_like(x2w)
    aa = [3,3.5,4, 4.5]
    for i, a in enumerate(aa):
        upper_lim = wcu.uppper_lim_with_wire(x1*wcu.s2_width_model(100e3,*par),
                                x1*wcu.f_upper_boundary(a,*fit_par[1]),
                                             x1*10**a * 0.75, x2w)/1e3
        if i == 0: data['uppper_lim_with_wire_1'] = upper_lim
        if i == 1: data['uppper_lim_with_wire_2'] = upper_lim
        if i == 2: data['uppper_lim_with_wire_3'] = upper_lim
        if i == 3: data['uppper_lim_with_wire_4'] = upper_lim
                                             
        #plt.plot(x2w,upper_lim,ls=':',
        #     label=f'cut boundary with S2=10^({a}) PE, dt=100 $\mu$s')
    #plt.ylabel('S2 width - model ($\mu$s)', ha='right', y=1)
    #plt.xlabel("distance from wire (cm)", ha='right', x=1)
    #plt.legend(loc='upper right')
    if name is not None: data.to_pickle(name)
    return data
```

```python
def save_events(events, sim=False, cut_mask=None, name = None):
    columns = ('cs1',
               'cs2',
               'e_ces',
               's2_area',
               's2_area_fraction_top',
               's2_range_50p_area',
               'drift_time',
               's2_normalized_width',
               's2_x_mlp',
               's2_y_mlp',
               'x_to_wire',
               'diff_width',
               'cut_near_wires',
               'cut_s2_width_wire_modeled_low_er')
    s2width = events['s2_range_50p_area']
    drift = events['drift_time']
    vd = get_correction_from_cmt('024075',('electron_drift_velocity', 'ONLINE', True))
    gd = get_correction_from_cmt('024075',('electron_drift_time_gate', 'ONLINE', True))
    dc = get_correction_from_cmt('024075',('electron_diffusion_cte', 'ONLINE', True))
    par = (dc,vd,gd)
    normWidth= wcu.s2_width_norm(s2width,wcu.s2_width_model(drift, *par))
    if not sim:
        x_to_wire = wcu.get_x_to_wire(events)
        diff_width = wcu.get_diff_width(events)
    
    data = pd.DataFrame(columns=columns)
    for column in columns:
        if column == 'cut_s2_width_wire_modeled_low_er':
            if sim:
                data[column] = cut_mask
            else: data[column] = events[column]
        if column == 's2_normalized_width': data[column] = normWidth
        elif column == 'x_to_wire': data[column] = x_to_wire
        elif column == 'diff_width': data[column] = diff_width
        else: data[column] = events[column]
    if name is not None: data.to_pickle(name)
    return data
```
