```python
import matplotlib.pyplot as plt
def plot_peak(p, t0=None, **kwargs):
    n = p['length']
    if t0 is None:
        t0 = p['time']
    plt.plot((p['time'] - t0) + np.arange(n) * p['dt'], p['data'][:n] / p['dt'], drawstyle='steps-mid',**kwargs)
    plt.xlabel("Time (ns)")
    plt.ylabel("Sum waveform (PE / ns)")
    plt.figure(2)
    plt.plot((p['time'] - t0)+ np.arange(n), p['data'][:n], drawstyle='steps-mid',color='red', **kwargs)
    plt.xlabel("Samples")
    plt.ylabel("Sum waveform (PE / sample)")
plot_peak(peaks[4])
```
