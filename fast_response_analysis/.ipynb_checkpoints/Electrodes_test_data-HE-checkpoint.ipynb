```python
def draw_wave(data, ev):
    length_wv= len(data['data'][ev])
    area=data['data'][ev].sum()
    t_wv=np.linspace(0,550,length_wv)
    plt.step(t_wv,data['data'][ev])
    plt.title("Run: %s Area:%.2f Time_ev: %f"%(run_name,area,(data['time'][ev]-data['time'][0])/1e9) )
```

```python
def set_cut_area_50p(data,x_m,x_M,y_m,y_M,c):
    '''
    data: full dataset
    x_m,x_M: range of var_x cut
    y_m,y_M: range of var_y cut
    c: color string for the lines    
    '''
    plt.figure(figsize=(10,8))
    hh=Histdd(np.log10(data['area']),np.log10(data['width'][:,5]),bins=(100,100),range=[(0,5),(0,4)],
             axis_names=('$\log_{10}(area)$ [PE]','$\log_{10}(range\_50p\_area)$ [ns]'))
    hh.plot('log')
    x_rng = [x_m,x_M]#[1.7,2.5]
    y_rng = [y_m,y_M]#[2.5,2.95]
    plt.vlines(x_rng[0],y_rng[0],y_rng[1],colors=c)
    plt.hlines(y_rng[0],x_rng[0],x_rng[1],colors=c)
    plt.vlines(x_rng[1],y_rng[0],y_rng[1],colors=c)
    plt.hlines(y_rng[1],x_rng[0],x_rng[1],colors=c)
    
    #cut = data['n_channels'] > 3;
    cut = (np.log10(data['area'])>x_rng[0]) & (np.log10(data['area'])<x_rng[1]);
    cut &= (np.log10(data['width'][:,5])>y_rng[0]) & (np.log10(data['width'][:,5])<y_rng[1]);
    return data[cut];
```
