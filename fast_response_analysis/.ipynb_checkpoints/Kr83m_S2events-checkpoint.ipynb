```python
def plot_area_width(run_id, low = 0, high = 5, low2 = 1, high2 = 5, binning = 500):
    events = st.get_array(run_id,'event_basics')
    ph_s1 = Histdd(events['s1_area'], events['s1_range_50p_area'],
                    bins=(np.logspace(low, high, binning), np.logspace(low2, high2, binning)))
    ph_s2 = Histdd(events['s2_area'], events['s2_range_50p_area'],
                    bins=(np.logspace(low, high, binning), np.logspace(low2, high2, binning)))
    plt.figure(figsize=(12,6))
    ph_s1.plot(log_scale=True, cblabel='S1 events',cmap='plasma')
    ph_s2.plot(log_scale=True, cblabel='S2 events')
    plt.xlabel("peak area (PE)", ha='right', x=1)
    plt.ylabel("peak width 50% (ns)", ha='right', y=1)
    plt.title(f'run {run_id}')
    plt.xscale('log')
    plt.yscale('log')
```

```python
plot_area_width('011669')
```

```python
plot_area_width('011673')
```

```python
def drift_velocity(run_id, low = 0, high = 6, binning = 500):
    events = st.get_array(run_id,'event_basics')
    plt.figure(figsize=(12,6))
    hdtime = Hist1d(events['drift_time']/1e6, bins=(np.linspace(low, high, binning)))
    hdtime.plot()
    plt.ylabel("events", ha='right', y=1)
    plt.xlabel("drift time (ms)", ha='right', x=1)
    plt.yscale('log')
```

```python
drift_velocity('011673')
```
